```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
tau = 1
rho = 1
L = 1
v0 = 1
```

```python
def w_r(r):
    
    return r*np.pi*np.sqrt(tau/rho)/L


def nu_r(r):
    
    return (    - (12 * (np.sin(np.pi*r/6))) / (np.power(np.pi,3)*np.power(r,3))
                - (12 * (np.sin(np.pi*r/3))) / (np.power(np.pi,3)*np.power(r,3))
                + (12 * (np.sin(2*np.pi*r/3))) / (np.power(np.pi,3)*np.power(r,3))
                + (12 * (np.sin(5*np.pi*r/6))) / (np.power(np.pi,3)*np.power(r,3))
                - (12 * (np.sin(np.pi*r)))  / (np.power(np.pi,3)*np.power(r,3))      ) 


def q_r(x, t, r):
    
    return -nu_r(r) * np.sin(w_r(r)*t) * np.sin(r*np.pi*x/L)


def qdot_r(x, t, r):
    
    return w_r(r) * -nu_r(r) * np.cos(w_r(r)*t) * np.sin(r*np.pi*x/L)
   


def q(x, t, R):
    
    suma = 0
    for r in range(1,R+1):
        suma += q_r(x, t, r)
    
    return suma


def qdot(x, t, R):
    
    suma = 0
    for r in range(1,R+1):
        suma += qdot_r(x, t, r)
        
    return suma
```

# Modo No Nulo 1

```python
x = np.linspace(0,L,100)

fig, ax = plt.subplots()

line2 = ax.plot(x, qdot_r(x,0,2))[0]

ax.set(xlim=[0, L], xlabel='x', ylabel='q (x,0)', title='velocidad en t=0 para r = 2')

plt.show()
```

# Animación con 5000 modos

```python
x = np.linspace(0,L,100)
t = np.linspace(0,2.5,100)

fig, ax = plt.subplots()

line2 = ax.plot(x, q(x,0,5000))[0]
ax.set(xlim=[0, L], ylim=[-0.8, 0.8], xlabel='x', ylabel='q (x,0)')

def update(frame):
    line2.set_data(x, q(x,t[frame],5000))
    ax.set_title(f"Evolución Temporal: time={t[frame]:.3f} s")
    return line2


ani = animation.FuncAnimation(fig=fig, func=update, frames=100, interval=30)
ani.save('Cuerda.gif')
ani.save('Cuerda.mp4')
plt.show()
```
